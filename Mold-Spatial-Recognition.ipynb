## Business Problem

Our client dGFabRC is a custom fabrication shop which works in a variety of materials and processes including millwork, metalwork, and concrete. They currently have a robotic arm that assists them in each of their separate types of work in the shop. For millwork, the robot is utilized for machining large objects from blocks of solid wood or Cross Laminated Timber. For metalwork, the robot is used for welding tube and sheet metal parts together for weldments. For concrete, the robot sprays GFRC (Glass Fiber Reinforced Concrete) into molds to create a part. 

In all of these instances, the robot must locate the work piece as it relates to the position of the robot. In the setup of the robotic workcell, a workpiece is required to be aligned on a positioning table in order to have the absolute position of the piece as it relates to the robot's position. In instances where the workpiece is large, the table is extremely large and immovable and the loading and unloading of material can take upwards of an hour to have done.

The robotic workcell has four sections to work in. They would like to integrate a solution to allow for fleixbility in all four sections of the robot's environment. In order to do so, they want to do away with immovable positioning tables and would like to setup a workpiece on a rolling table or rolling A-frame outside of the robotic workcell and wheel the piece into one section of the workcell while the robot is working in the other section.

In order for this to happen, the robot needs to have a computer vision system which can recognize whether the workpiece is in the section that the robot is looking at or not. Then if the workpiece is in fact in the robot's section, the algorithm must determine the spatial 3D position of the work piece. In this case, we will focus on the millwork process and recognize a block of wood to be machined.
