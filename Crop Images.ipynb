# Crop Images

```python
# Input original dir, new directory, and desired image size in below. Original dir should contain all of the images you would like to crop

original_dir =r'C:\Users\Tim\Documents\Flatiron\Phase_4\Material-Identification\images\Madjek_NoClass'
new_dir = r"C:\Users\Tim\Documents\Flatiron\Phase_4\Material-Identification\images\Madjek_Crop"
img_height = 512
img_width = 284
```

```python
# import packages
from PIL import Image
import os.path, sys
```

```python
# list the directory
dirs = os.listdir(original_dir)

# define the function to crop each image and save in new directory

def crop():
    for item in dirs:
        fullpath = os.path.join(original_dir,item)
        if os.path.isfile(fullpath):
            im = Image.open(fullpath)
            width, height = im.size
            left = (width - img_width)/2
            top = (height - img_height)/2
            right = (width + img_width)/2
            bottom = (height + img_height)/2
            imCrop = im.crop((left, top, right, bottom))
            
            if not os.path.exists(new_dir):
                os.mkdir(new_dir)
            src_path = original_dir + '/'+ item
            dst_path = new_dir + '/' + item
            
            try:
                imCrop.save(new_dir + '/' + item)
                print(item + ": sucessful")
            except IOError as e:
                print('Unable to save file {} to {}'
                      .format(src_path, dst_path))
            except:
                print('When try save file {} to {}, unexpected error: {}'
                      .format(src_path, dst_path, sys.exc_info()))

crop()
```
